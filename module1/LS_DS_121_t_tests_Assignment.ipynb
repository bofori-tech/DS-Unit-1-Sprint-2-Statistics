### 2) Pretend that this dataset represents the cars at a used car lot in your local town. 

- Use df.sample() to pick a random sample of 10 cars. Note that because this sample is random we are going to set the `random_state` so that all of us in the class get the same random sample. Please set your random state to `30` when using `df.sample()`

Is your sample reflective of the population value in regards to highway-mpg? Find the mean for `highway-mpg` for the entire dataset and use a 1-sample t-test to compare your estaimated sample mean to the population mean. Can you say that they are different? 

```python
df.sample(30)
```

```python
random_state=df.sample(30)
random_state.head()
```

```python
from scipy.stats import ttest_ind
random_state['highway-mpg'].mean()
df['highway-mpg'].mean()
print(random_state['highway-mpg'].mean())
print(df['highway-mpg'].mean())
```

```python
ttest_ind(random_state['highway-mpg'], df['highway-mpg'], nan_policy='omit')
```

The salesman says the cars he sells typically have a fuel efficiency of about 35 miles per gallon on the highway. You want to verify his claim but can only test 10 cars. Using your sample of 10, test his claim and report your results.

```python
df2=random_state.sample(10)
df2.head()
```

```python
from scipy.stats import ttest_1samp
popmean=df['highway-mpg'].mean()
print(popmean)
```

```python
# is the average fuel efficiency of random_state above the population average? yes.
ttest_1samp(df2['highway-mpg'], popmean, nan_policy='omit')
```

```python
#pvalue is greater the 0,therefore we reject the null hypothesis
#that 
```

Imagine that you now have the capacity to test 100 cars on the lot. Using the same random state of `30`, take a sample of 100 cars. Run a t-test to verify the salesman's claim with your new larger sample. Do you reach the same conclusion as you did with the sample size of 10? 

```python
df3=df.sample(100)
df3.head()
```

```python
ttest_1samp(df3['highway-mpg'], popmean, nan_policy='omit')
```

Why might these two t-tests using the same dataset lead to different conclusions about the salesman's claim?

### 2) Perform two sample T-tests on different issues and report the results.

- Find an issue that democrats support more than republicans with p < 0.01 (significant at the 99% level).
- Find an issue that republicans support more than democrats with p < 0.01 (significant at the 99% level).
- Find an issue where the difference between republicans and democrats has p > 0.1 (Not significant at the 90% level - i.e. there may not be much of a difference the two sample means)

Please for each test that you run state your null and alternative hypothesis and then write a conclusion reflecting on the null and alternative hypothesis.

Remember, that two-sample t-tests will only tell us if the two groups are *different* from one another. We'll have to look at their sample means directly or use the sign on the t-statistic to know which of the two sample means is larger. 

```python
from scipy.stats import ttest_ind
```

```python
#Average rate of voting 'yes' on handicapped-infants
dem['handicapped-infants'].mean()
```

```python
rep['handicapped-infants'].mean()
```

```python
#compare with t-test:
ttest_ind(dem['handicapped-infants'], rep['handicapped-infants'], nan_policy='omit')
```

```python
#I fail to accept the null hypothesis if my p-value is < .01 or if my p-value is less than (1-confidence_level)

Conclusion: due to a p-value of 1.613-18 I accept the null hypothesis that republican and democrat support for 
the handicapped-infants bill is different.

```

```python
#average rate of voting "yes" on immigration
dem['immigration'].mean()
```

```python
#average rate of voting 'yes' on immigration
rep['immigration'].mean()
```

```python
#compare with t-test:
ttest_ind(dem['immigration'], rep['immigration'], nan_policy='omit')
```

```python
#with a pvalue > .01 I fail to accept the null hypothesis than repulicans and democrats 
# support for the immigration bill is different
```

```python
#average rate of voting 'yes' on mx-missile
dem['education-spending'].mean()
```

```python
rep['education-spending'].mean()
```

```python
#compare with t-test:
ttest_ind(dem['education-spending'], rep['education-spending'], nan_policy='omit')
```

```python
#pvalue is < .01,I accept the null hypothesis that the democrats and republican support for education spending is different.
```
